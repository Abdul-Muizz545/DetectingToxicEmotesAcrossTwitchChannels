The code in this file explores the distribution for number of words in HasanAbi comments and how they are populated with emotes

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
import statistics
```

```python
"""Read csv file"""

#only need comment column
df = pd.read_csv("path to HasanAbi csv file", usecols = ["comment"])
df = df.dropna() #remove rows with null values
```

# Histogram for message length

```python
comment_length = [] #list to store length of each comment
for msg in df["comment"]:
  comment_length.append(len(msg.split(" ")))
print(comment_length)
```

Descriptive statistics for length of comments in words

```python
mean_comment_length = statistics.mean(comment_length)
median_comment_length = statistics.median(comment_length)
stddev_comment_length = statistics.stdev(comment_length)


print(f"Mean: {mean_comment_length}")
print(f"Median: {median_comment_length}")
print(f"Standard Deviation: {stddev_comment_length}")
```

```python
print("Number of HasanAbi comments: ", len(comment_length))
```

```python
comment_length = Counter(comment_length) #count number of times for each comment length value 

print(comment_length.items())
```

```python
comment_length_dictionary = { word: occurrences for word, occurrences in comment_length.items() if word <= 25 }
```

```python
print(comment_length_dictionary)
```

```python
# Plotting histogram for length of HasanAbi messages in words
plt.figure(figsize=(25,10))
plt.title('Histogram for HasanAbi comment length in words', fontsize = 20)
plt.margins(x=0)
plt.xticks(range(1, max(comment_length_dictionary.keys()) + 1))
plt.xlabel('Word Count', fontsize = 15)
plt.ylabel('Frequency', fontsize = 15)
bar = plt.bar(comment_length_dictionary.keys(), comment_length_dictionary.values())
plt.bar_label(bar)
plt.show()
```

# Histogram for number of emotes per message

```python
global_emotes_df = pd.read_csv("global_emotes.csv")
print(global_emotes_df)

print(global_emotes_df.isnull().sum())

# Convert DataFrame to list of strings, ignoring NaN values
list_of_global_emotes = []

for col in global_emotes_df.columns:
    column_data = global_emotes_df[col].dropna().astype(str).tolist()
    list_of_global_emotes.extend(column_data)
```

```python
print(list_of_global_emotes)
```

```python
channel_emotes_df = pd.read_csv("channel_emotes.csv")
#Renaming columns that start with numbers to avoid potential conflict with pandas
channel_emotes_df.columns = ['Col_' + str(col) if col[0].isdigit() else col for col in channel_emotes_df.columns]
```

```python
# Function that returns whether the word is a global emote, channel emote or not an emote
def wordIsChannelOrGlobalEmote(word, channel_emotes_df, global_emotes_list):
    column_with_word = []

    channel = "HasanAbi" #only considering HasanAbi channel for ease of time

    #for channel in channel_emotes_df.columns:
    if (channel_emotes_df[channel] == word).any() and word != "":
        column_with_word.append(channel)

    if word in global_emotes_list:
        return "global" #return that its global
    elif len(column_with_word) > 0:
        return column_with_word[0].removeprefix("Col_") #return channel name
    else:
        return "No" #not an emote
```

```python
num_emotes_per_message = [] #list to store number of emotes found in each message
counter = 0

for idx, row in df.iterrows(): #go through each msg in comment
    wordsList = row["comment"].split(" ") #Get list of words from each comment
    num_emotes_per_message.append(0)

    for word in wordsList:
      wordIsEmote = wordIsChannelOrGlobalEmote(word, channel_emotes_df, list_of_global_emotes)

      if wordIsEmote != "No": #word is an emote
        num_emotes_per_message[idx] += 1

    counter += 1

    if counter % 1000 == 0:
        print(str(counter) + " rows have been processed")
```

```python
num_emotes_per_message_counter = Counter(num_emotes_per_message) #count number of times each number appears

print(num_emotes_per_message_counter.items())
```

```python
num_emotes_per_message_dict = { word: occurrences for word, occurrences in num_emotes_per_message_counter.items() }
```

```python
# Plotting histogram for length of messages in words
plt.figure(figsize=(25,10))
plt.title('Histogram for number of emotes per message', fontsize = 20)
plt.margins(x=0)
# plt.xticks(rotation=90)
plt.xticks(range(0, max(num_emotes_per_message_counter.keys()) + 1))
plt.xlabel('Number of emotes in message', fontsize = 15)
plt.ylabel('Frequency', fontsize = 15)
bar = plt.bar(num_emotes_per_message_dict.keys(), num_emotes_per_message_dict.values())
plt.bar_label(bar)
plt.show()
```

# Histogram for ratio of number of emotes over message length (to show that emotes populate messages)

```python
msg_length_list = [] #list to store length of each message in words


for idx, row in df.iterrows(): #go through each msg in comment
    msg_length = len(row["comment"].split(" ")) #Get list of words from each comment
    msg_length_list.append(msg_length)

print(msg_length_list)
```

```python
emote_message_ratio_list = []

for idx in range(0, len(msg_length_list)):
  ratio = num_emotes_per_message[idx] / msg_length_list[idx]

  emote_message_ratio_list.append(round(ratio, 2))
```

```python
print(emote_message_ratio_list)
```

```python
emote_message_ratio_counter = Counter(emote_message_ratio_list)

print(emote_message_ratio_counter.items())
```

```python
# Plot the histogram
heights1, bin_edges1, patches1 = plt.hist(emote_message_ratio_list, bins = int((max(emote_message_ratio_list) - min(emote_message_ratio_list)) / 0.1), color = "blue")
plt.title('Histogram of emote vs message length ratio')
plt.xlabel('Ratio of number of emotes and message length')
plt.ylabel('Frequency')

plt.grid(True)
plt.show()
```
