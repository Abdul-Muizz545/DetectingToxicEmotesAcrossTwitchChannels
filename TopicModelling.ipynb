```python
import pandas as pd
from top2vec import Top2Vec
```

```python
#In this case we chose two of HasanAbi's chat logs to analyze, but you can choose any number of chat logs and then combine them into one large dataframe

#only need the comment column
df1 = pd.read_csv("path to HasanAbi csv file ", usecols = ["comment"])
df1 = df1.dropna() #remove rows with null values

df2 = pd.read_csv('path to HasanAbi csv file', usecols = ["comment"])
df2 = df2.dropna()

```

```python
df = pd.concat([df1, df2], ignore_index = True)
df
```

```python
text = df["comment"].tolist()
```

```python
model = Top2Vec(documents = text) #Each document is a comment in the Twitch stream, so many commments means many documents
```

# Save and load model

```python
model.save("Top2Vecmodel.h5")
```

```python
import keras
model = keras.models.load_model('Top2Vecmodel.h5')
```

# Extracting topics in HasanAbi's channel

```python
topic_sizes, topic_nums = model.get_topic_sizes()

print(topic_sizes) #Number of comments most similar to each of the topics we found (in descending order)
```

```python
print(model.get_num_topics())
```

```python
topic_words, word_scores, topic_nums = model.get_topics(100) #Getting the top 100 topics

#topic_words is list of top 50 words semantically similar to each topic
#word_scores is cosine similarity of top 50 words to the topic at hand
```

```python
for words, scores, num in zip(topic_words, word_scores, topic_nums):
  print(num)
  print(words)
```
